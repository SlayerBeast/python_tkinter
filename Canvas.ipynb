# 19-1 绘图功能

## 19-1-1 建立画布
- 用Canvas()方法建立画布对象
```
tk=Tk()          #使用tk当窗口Tk对象
canvas=Canvas(tk,width=xx,height=yy)       #xx,yy是画布宽与高
canvas.pack()                              #将画布包装好
```
- 画布建立完成后，左上角是坐标(0,0)，向右x轴递增，向下y轴递增

## 19-1-2 绘制线条create_line()
- create_line(x1,y1,x2,y2,...,xn,yn,options),线条会沿着(x1,y1),(x2,y2),...绘制下去。create_line()参数至少有2组坐标。
 - arrow:默认没有箭头，使用arrow=tk.FIRST在起始线末端有箭头，arrow=LAST在最后一条线末端有箭头，arrow=tk.BOTH在两端有箭头。
 - arrowshape:使用元组(d1,d2,d3)代表箭头，默认是(8,10,3).
 - capstyle:线条终点的样式，默认是BUTT，也可选择PROJECTING，ROUND。
 - dash:建立虚线。使用元组储存数字数据，第一个数字是实线，第二个数字是空白，如此循环当所有元组数字用完又重新开始。如dash=(5,3)产生5像素实线，3像素空白，如此循环。如dash=(8,1,1,1)产生8像素实线和点的线条，dash=(5,)产生5像素实线和5像素空白。
 - dashoffset:与dash一样产生虚线，但是一开始数字是空白的宽度。
 - fill:设置线条颜色。
 - joinstyle:线条相交的设置，默认是ROUND，也可选择BEVEL，MITER。
 - stipple:绘制位图(Bitmap)线条。
 - width:线条宽度。

````python
## 在半径为100的圆外围建立12个点，然后将这些点彼此连接。

from tkinter import *
import math

tk=Tk()
canvas=Canvas(tk,width=640,height=480)
canvas.pack()
x_center,y_center,r=320,240,100
x,y=[],[]
for i in range(12):                         #建立圆外围12个点
    x.append(x_center+r*math.cos(30*i*math.pi/180))
    y.append(y_center+r*math.sin(30*i*math.pi/180))
for i in range(12):                         #将12个点彼此连线
    for j in range(12):
        canvas.create_line(x[i],y[i],x[j],y[j])
tk.mainloop()
````

- create_rectangle(x1,y1,x2,y2,options),(x1,y1)和(x2,y2)是矩形左上角和右下角的坐标。
 - dash:建立虚线，与create_line()相同
 - dashoffset:与dash一样产生虚线，但开始数字是空白的宽度
 - fill:矩形填充颜色
 - outline:设置矩形线条颜色

````python
#在画布内随机产生不同位置与大小的矩形

from tkinter import *
from random import *

tk=Tk()
canvas=Canvas(tk,width=640,height=480)
canvas.pack()
for i in range(50):                              #随机绘制50个不同位置与大小的矩形
    x1,y1=randint(1,640),randint(1,480)
    x2,y2=randint(1,640),randint(1,480)
    if x1>x2:x1,x2=x2,x1                         #确保左上角x坐标小于右下角x坐标
    if y1>y2:y1,y2=y2,y1                         #确保左上角y坐标小于右下角y坐标
    canvas.create_rectangle(x1,y1,x2,y2,fill="blue",outline="red",width=10)

tk.mainloop()
````

## 19-1-4 绘制圆弧create_arc()
- create_arc(x1,y1,x2,y2,extent=angle,style=ARC,options)
- (x1,y1)和(x2,y2)分别是圆弧左上角和右下角的坐标。
- options:
 - dash:建立虚线，与create_line()相同。

 - extent:若要绘制原型extent值是359，若写成360会被视为0度。若extent介于1～359，则是绘制这个角度的圆弧。
 - fill:填充圆弧颜色。
 - outline:设置圆弧线条颜色。
 - start:圆弧起点位置
 - stipple:绘制位图圆弧
 - style:有三种格式——ARC，CHORD，PIESLICE
 - width:圆弧线条宽度
 - style=ARC表示绘制圆弧，若要使用options参数填满圆弧，则需舍去此参数。
 - 默认绘制圆弧的起点是右边，也可用start=0代表，也可设置start的值更改圆弧的起点，方向是逆时针。

````python
#绘制各种不同的圆和椭圆，圆弧，椭圆弧。

from tkinter import *

tk=Tk()
canvas=Canvas(tk,width=640,height=480)
canvas.pack()
#以下以圆形为基础
canvas.create_arc(10,10,110,110,extent=45,style=ARC)
canvas.create_arc(210,10,310,110,extent=90,style=ARC)
canvas.create_arc(410,10,510,110,extent=180,fill="yellow")
canvas.create_arc(10,110,110,210,extent=270,style=PIESLICE)
canvas.create_arc(210,110,310,210,extent=359,style=ARC,width=5)
#以下以椭圆形为基础
canvas.create_arc(10,250,310,350,extent=90,style=ARC,start=90)
canvas.create_arc(320,250,620,350,extent=180,style=ARC)
canvas.create_arc(10,360,310,460,extent=270,style=ARC,outline="blue")
canvas.create_arc(320,260,620,460,extent=359,style=ARC)

tk.mainloop()
````

# 19-2 鼠标拖曳绘制线条
- 使用鼠标拖曳时绑定paint事件处理程序，在事件中取得鼠标坐标，然后用create_oval()方法绘制极小化的圆，方法是圆的左上角坐标与右下角坐标相同。
- canvas.delete("all")，可以清除画布。

````python
#设计一个简单的绘图程序，该程序在执行时若拖曳鼠标可以绘制线条

from tkinter import *

def paint(event):                                #拖曳可以绘图
    x1,y1=(event.x,event.y)                      #设置左上角坐标
    x2,y2=(event.x,event.y)                      #设置右下角坐标
    canvas.create_oval(x1,y1,x2,y2,fill="blue")
    
def cls():                                       #清除画面
    canvas.delete("all")
    
tk=Tk()
lab=Label(tk,text="拖曳鼠标可以绘图")               #建立标题
lab.pack()
canvas=Canvas(tk,width=640,height=300)           #建立画布
canvas.pack()

btn=Button(tk,text="清除",command=cls)            #建立“清除”按钮
btn.pack(pady=5)

canvas.bind("<B1-Motion>",paint)                 #鼠标拖曳绑定paint

canvas.mainloop()
````

# 19-3 动画设计

## 19-3-1 基本动画
- 动画设计所使用的方法是move()，格式如下：
```
canvas.move(ID,xMove,yMove)                #ID是对象编号
canvas.update()                            #强制重绘画布
```
- xMove,yMove分别是沿x和y轴移动距离，单位是像素

````python
#移动球的设计，每次移动5像素

from tkinter import *
import time

tk=Tk()
canvas=Canvas(tk,width=500,height=150)
canvas.pack()
canvas.create_oval(10,50,60,100,fill="yellow",outline="lightgray")
for x in range(0,80):
    canvas.move(1,5,0)                             #ID=1,x轴移动5像素，y轴不变
    canvas.update()                                #强制Tkinter 重绘
    time.sleep(0.05)

tk.mainloop()
````

````python
#一次移动两个球，黄色球是id1，水蓝色球是id2

from tkinter import *
import time

tk=Tk()
canvas=Canvas(tk,width=500,height=250)
canvas.pack()
id1=canvas.create_oval(10,50,60,100,fill="yellow")
id2=canvas.create_oval(10,150,60,210,fill="aqua")
for x in range(0,80):
    canvas.move(id1,5,0)                          #id1 x轴移动5像素，y轴移动0像素
    canvas.move(id2,5,0)                          #id2 x轴移动5像素，y轴移动0像素
    tk.update()                                   #强制Tkinter重绘
    time.sleep(0.05)
    
tk.mainloop()
````

## 19-3-3 将随机数应用于多个球体的移动

````python
#循环跑100次看哪一个球跑得快，让黄色球有70%赢的机会。

from tkinter import *
import time

tk=Tk()
canvas=Canvas(tk,width=500,height=250)
canvas.pack()
id1=canvas.create_oval(10,50,60,100,fill="yellow")
id2=canvas.create_oval(10,150,60,210,fill="aqua")
for x in range(0,100):
    if randint(1,100)>70:
        canvas.move(id2,5,0)                          #id2 x轴移动5像素，y轴移动0像素
    else:
        canvas.move(id1,5,0)                          #id1 x轴移动5像素，y轴移动0像素
    tk.update()                                   #强制Tkinter重绘
    time.sleep(0.05)
    
tk.mainloop()
````

## 19-3-4 消息绑定
- 主要思路是利用系统接收到键盘的消息，做出反应。
- 当发生按下右移键时，可以控制球往右边移动
```
def ballMove(event):
    canvas.move(1,5,0)                  #假设移动5像素
```
- 对于按下右移键移动球可以这样设计
```
def ballMove(event):
    if event.keysym=="Right":
        canvas.move(1,5,0)
```
- 对于主程序而言需使用canvas.bind_all()函数，执行消息绑定工作，代码如下
```
canvas.bind_all("<KeyPress-Left>",ballMove)       #左移键
canvas.bind_all("<KeyPress-Right>",ballMove)      #右移键
canvas.bind_all("<KeyPress-Up>",ballMove)         #上移键
canvas.bind_all("<KeyPress-Down>",ballMove)       #下移键
```

````python
#画布中央有一个红球，按键盘上的上下左右键，可往上下左右移动球，每次移动5个像素

from tkinter import *
import time

def ballMove(event):
    if event.keysym=="Left":     #左移
        canvas.move(1,-5,0)
    if event.keysym=="Right":    #右移
        canvas.move(1,5,0)
    if event.keysym=="Up":       #上移
        canvas.move(1,0,-5)
    if event.keysym=="Down":     #下移
        canvas.move(1,0,5)
        
tk=Tk()
canvas=Canvas(tk,width=500,height=300)
canvas.pack()
canvas.create_oval(225,125,275,175,fill="red")
canvas.bind_all("<KeyPress-Left>",ballMove)
canvas.bind_all("<KeyPress-Right>",ballMove)
canvas.bind_all("<KeyPress-Up>",ballMove)
canvas.bind_all("<KeyPress-Down>",ballMove)
mainloop()
````

# 19-4 反弹球游戏设计
- 本节介绍设计一个反弹球的游戏

## 19-4-1 设计球往下移动

````python
#定义画布宽度与高度分别为640像素，480像素。这个球将往下移动然后消失，移到超出画布范围就消失了。

from tkinter import *
from random import *
import time

class Ball:
    def __init__(self,canvas,color,winW,winH):
        self.canvas=canvas
        self.id=canvas.create_oval(0,0,20,20,fill=color)          #建立球对象
        self.canvas.move(self.id,winW/2,winH/2)                   #设置球最初位置
    def ballMove(self):
        self.canvas.move(self.id,0,step)                          #step是正值表示往下移动
        
winW=640                                                          #定义画布宽度
winH=480                                                          #定义画布高度
step=3                                                            #定义速度，可想成位移步长
speed=0.03                                                        #设置移动速度

tk=Tk()
tk.title("Bouncing Ball")                                         #游戏窗口标题
tk.wm_attributes("-topmost",1)                                    #确保游戏窗口在屏幕最上层
canvas=Canvas(tk,width=winW,height=winH)
canvas.pack()
tk.update()

ball=Ball(canvas,"yellow",winW,winH)                              #定义球对象

while True:
    ball.ballMove()
    tk.update()
    time.sleep(speed)                                             #可以控制移动速度
mainloop()
````

## 19-4-2 设计让球上下反弹
- 可用coords()方法获得对象位置，返回值是对象的左上角和右下角坐标。

````python
#建立一个球，用coords()方法列出球位置的消息

from tkinter import *
tk=Tk()
canvas=Canvas(tk,width=500,height=150)
canvas.pack()
id=canvas.create_oval(10,50,60,100,fill="yellow",outline="lightgray")
ballPos=canvas.coords(id)
print(ballPos)
````

- 从上例结果看，由ballPos=canvas.coords(id)可得：
 - ballPos[0]:球的左边x轴坐标，可用于判断是否撞到画布左方
 - ballPos[1]:球的上边y轴坐标，可用于判断是否撞到画布上方
 - ballPos[2]:球的右边x轴坐标，可用于判断是否撞到画布右方
 - ballPos[3]:球的下边y轴坐标，可用于判断是否撞到画布下方

````python
#设计让球可以上下移动。

from tkinter import *
from random import *
import time

class Ball:
    def __init__(self,canvas,color,winW,winH):
        self.canvas=canvas
        self.id=canvas.create_oval(0,0,20,20,fill=color)          #建立球对象
        self.canvas.move(self.id,winW/2,winH/2)                   #设置球最初位置
        self.x=0                                                  #水平不移动
        self.y=step                                               #垂直移动单位
    def ballMove(self):
        self.canvas.move(self.id,self.x,self.y)                   
        ballPos=self.canvas.coords(self.id)
        if ballPos[1]<=0:                                         #侦测球是否超过画布上方
            self.y=step
        if ballPos[3]>=winH:                                      #侦测球是否超过画布下方
            self.y=-step
        
winW=640                                                          #定义画布宽度
winH=480                                                          #定义画布高度
step=3                                                            #定义速度，可想成位移步长
speed=0.03                                                        #设置移动速度

tk=Tk()
tk.title("Bouncing Ball")                                         #游戏窗口标题
tk.wm_attributes("-topmost",1)                                    #确保游戏窗口在屏幕最上层
canvas=Canvas(tk,width=winW,height=winH)
canvas.pack()
tk.update()

ball=Ball(canvas,"yellow",winW,winH)                              #定义球对象

while True:
    ball.ballMove()
    tk.update()
    time.sleep(speed)                                             #可以控制移动速度
mainloop()
````

## 19-4-3 设计让球在画布四面反弹
- 考虑到球会在x轴移动，前例中Ball类别的__init__()函数需从
```
self.x=0             #水平不移动
self.y=step          #垂直移动单位
```
- 修改为：
```
startPos=[-4,-3,-2,-1,1,2,3,4]                   #球最初x轴位移的随机数
shuffle(startPos)                                #打乱排列
self.x=startPos[0]                               #球最初水平移动单位
self.y=step                                      #垂直移动单位
```

- **shuffle(list)方法，可将list中的所有元素随机排序，需先导入random模块.**

````python
#完整的反弹球设计。

#增加设计让球拍左右可以移动。

from tkinter import *
from random import *
import time

class Ball:
    def __init__(self,canvas,color,winW,winH,racket):
        self.canvas=canvas
        self.racket=racket
        self.id=canvas.create_oval(0,0,20,20,fill=color)          #创建球对象
        self.canvas.move(self.id,winW/2,winH/2)                   #设置球最初位置
        startPos=[-4,-3,-2,-1,1,2,3,4]                            #球最初x轴位移的随机数
        shuffle(startPos)                                         #打乱排列
        self.x=startPos[0]                                        #球最初水平移动单位
        self.y=-step                                              #垂直移动单位
        self.notTouchBottom=True                                  #未接触画布底端
    def hitRacket(self,ballPos):
        racketPos=self.canvas.coords(self.racket.id)
        if ballPos[2]>=racketPos[0] and ballPos[0]<=racketPos[2]:
            if ballPos[3]>=racketPos[1] and ballPos[3]<=racketPos[3]:
                return True
        return False
    def ballMove(self):
        self.canvas.move(self.id,self.x,self.y)                   
        ballPos=self.canvas.coords(self.id)
        if ballPos[0]<=0:                                         #侦测球是否超过画布左方
            self.x=step
        if ballPos[1]<=0:                                         #侦测球是否超过画布上方
            self.y=step
        if ballPos[2]>=winW:                                      #侦测球是否超过画布右方
            self.x=-step
        if ballPos[3]>=winH:                                      #侦测球是否超过画布下方
            self.y=-step
        if self.hitRacket(ballPos)==True:                         #侦测是否撞到球拍
            self.y=-step
        if ballPos[3]>=winH:                                      #若球接触到画布底端
            self.notTouchBottom=False
        
class Racket:
    def __init__(self,canvas,color):
        self.canvas=canvas
        self.id=canvas.create_rectangle(0,0,100,15,fill=color)    #球拍对象
        self.canvas.move(self.id,270,400)                         #球拍位置
        self.x=0
        self.canvas.bind_all("<KeyPress-Right>",self.moveRight)   #绑定按住右键
        self.canvas.bind_all("<KeyPress-Left>",self.moveLeft)     #绑定按住左键
    def racketMove(self):                                         #设计球拍移动
        racketPos=self.canvas.coords(self.id)
        if racketPos[0]<=0 and self.x<=0:                         #移动时是否碰到画布左边
            self.x=0
        elif racketPos[2]>=winW and self.x>=0:                    #移动时是否碰到画布右边
            self.x=0
        self.canvas.move(self.id,self.x,0)
    def moveLeft(self,event):                                     #球拍每次向左移动的单位数
        self.x=-3
    def moveRight(self,event):                                    #球拍每次向右移动的单位数
        self.x=3
        

winW=640                                                          #定义画布宽度
winH=480                                                          #定义画布高度
step=3                                                            #定义速度，可想成位移步长
speed=0.01                                                        #设置移动速度

tk=Tk()
tk.title("Bouncing Ball")                                         #游戏窗口标题
tk.wm_attributes("-topmost",1)                                    #确保游戏窗口在屏幕最上层
canvas=Canvas(tk,width=winW,height=winH)
canvas.pack()
tk.update()

racket=Racket(canvas,"purple")                                    #定义紫色球拍
ball=Ball(canvas,"yellow",winW,winH,racket)                       #定义球对象

while ball.notTouchBottom:                                        #若球未接触画布底端
    try:
        ball.ballMove()
    except:
        print("单击关闭按钮终止程序执行")
        break
    racket.racketMove()
    tk.update()
    time.sleep(speed)                                             #可以控制移动速度
tk.mainloop()


#主程序循环中的捕捉异常方式处理语句使单击窗口“关闭”按钮时，不再有报错。
````
